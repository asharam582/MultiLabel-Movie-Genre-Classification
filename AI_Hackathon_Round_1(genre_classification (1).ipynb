```python
X = train_df['plot_synopsis']
```

```python
# X = train_df['plot_synopsis']
y= train_df.drop(['index', 'plot_synopsis'], axis=1)
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# Split the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.1, random_state=42)

# Vectorization
max_features = 20000
max_len = 1500
vectorizer = TextVectorization(max_tokens=max_features, output_sequence_length=max_len, output_mode='int')
vectorizer.adapt(X_train.values)
vectorized_text_train = vectorizer(X_train.values)
vectorized_text_val = vectorizer(X_val.values)

# Data Pipeline for training and validation sets
dataset_train = tf.data.Dataset.from_tensor_slices((vectorized_text_train, y_train.values))
dataset_train = dataset_train.cache().shuffle(160000).batch(16).prefetch(8)
dataset_val = tf.data.Dataset.from_tensor_slices((vectorized_text_val, y_val.values))
dataset_val = dataset_val.cache().batch(16).prefetch(8)

# Model Building
model = Sequential([
    Embedding(max_features + 1, 32),
    Bidirectional(LSTM(32, activation='tanh')),
    Dense(128, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(len(train_df.columns) - 2, activation='sigmoid')  # excluding 'index' and 'plot_synopsis'
])

model.compile(optimizer='Adam', loss='binary_crossentropy')

# Training
hist = model.fit(dataset_train, epochs=5, validation_data=dataset_val, verbose=1)

```

```python
# Prediction
vectorized_text_test = vectorizer(test['plot_synopsis'].values)
predictions = model.predict(vectorized_text_test)
```
